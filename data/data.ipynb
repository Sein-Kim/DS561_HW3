```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Data Load
urb_core = pd.read_csv('./SeoulData/urban-core_v2.csv')
urb_mix = pd.read_csv('./SeoulData/urban-mix_v2.csv')

'''
speeddata 열이름 지정 
형식 : 요MMDDHHMM
# 요 : 요일 1 2 3 4 5 6 7 (월 화 수 목 금 토 일)
# MM : 월 04 (4월)
# DD : 일 01~30 (1일~30일)
# HHMM : 시간 0000~2355 (00:00~23:55)

예시 : 일요일 4월 1일 13시 25분 = 704011325
'''
```

```python
#속도 data 추출 (2018/4/1 ~ 2018/4/30) 5분간격
speed_data_core = urb_core.drop(columns=['Link_ID_1'	,'Link_ID_2',	'Center_Point_1',	'Center_Point_2',	'Limit'	,'Length',	'Direction'])
speed_data_mix = urb_mix.drop(columns=['Link_ID_1'	,'Start_1',	'Start_2',	'End_1',	'End_2'	,'Limit',	'-'])

print(speed_data_core.shape)
print(speed_data_mix.shape)
speed_data_mix.head()
```

```python
adj_core = pd.read_csv('./SeoulData/Adj(urban-core).csv', header=None)
adj_mix = pd.read_csv('./SeoulData/Adj(urban-mix).csv', header=None)

print(adj_core.shape)
print(adj_mix.shape)
```

```python
import datetime

# Your original time format
time_str = "704010000"

def change_col_name(time_str) :
    # Extracting components
    day_of_week = int(time_str[0])
    month = int(time_str[1:3])
    day = int(time_str[3:5])
    hour = int(time_str[5:7])
    minute = int(time_str[7:9])

    # Find the corresponding Monday date in 2018 (for this example, April 2, 2018)
    monday_date = datetime.date(2018, 4, 2)

    # Calculate the date by adding the difference (considering day_of_week starts from 1 for Monday)
    actual_date = monday_date + datetime.timedelta(days=day_of_week - 1)

    # Replace month and day with the extracted values
    actual_date = actual_date.replace(month=month, day=day)

    # Format the final datetime string
    formatted_date = actual_date.strftime('%Y/%m/%d') + f" {hour:02d}:{minute:02d}:00"

    return formatted_date

change_col_name("704010000")
```

```python
col_names = []
for col_name in speed_data_core.columns:
    col = change_col_name(col_name)
    col_names.append(col)

speed_data_core.columns = col_names
speed_data_core.head()
```

```python
speed_data_core = speed_data_core.T
speed_data_core.index = pd.to_datetime(speed_data_core.index, format='%Y/%m/%d %H:%M:%S')
speed_data_core.to_csv('./SeoulData/urban_core_gcn.csv')
speed_data_core.to_hdf('./seoul-core.h5', key='df')
```
